```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Hàm mất mát
def loss(theta):
    return theta**2  # Ví dụ hàm mất mát là bình phương

# Đạo hàm của hàm mất mát
def gradient(theta):
    return 2 * theta  # Ví dụ đạo hàm là 2 * theta

# Tốc độ học
learning_rate = 0.1

# Tạo dãy giá trị theta
theta_values = np.linspace(-5, 5, 100)

# Tính giá trị hàm mất mát tại từng giá trị của theta
loss_values = loss(theta_values)

# Tính giá trị đạo hàm tại từng giá trị của theta
gradient_values = gradient(theta_values)

# Biểu diễn
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Vẽ hàm mất mát
ax.plot(theta_values, np.zeros_like(theta_values), loss_values, label='Loss Function')

# Vẽ gradient
ax.quiver(theta_values, np.zeros_like(theta_values), loss_values, np.zeros_like(theta_values), gradient_values, np.zeros_like(theta_values),
          length=0.1, normalize=True, color='r', label='Gradient')

# Vẽ tốc độ học
ax.text2D(0.05, 0.95, f'Learning Rate: {learning_rate}', transform=ax.transAxes, fontsize=12, color='blue')

ax.set_xlabel('Theta')
ax.set_ylabel('Learning Rate')
ax.set_zlabel('Loss')
ax.set_title('Biểu đồ 3D của Hàm Mất Mát, Gradient và Learning Rate')

plt.legend()
plt.show()
```
